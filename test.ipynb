```python
import tensorflow as tf
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load the saved model
model = tf.keras.models.load_model('D:/Leaf machine/model.h5')

def preprocess_image(image_path):
    image = Image.open(image_path).convert('RGB')
    image = image.resize((300, 300))
    image = np.array(image)
    image = image / 255.0  # Normalize the pixel values
    image = np.expand_dims(image, axis=0)  # Add an extra dimension for batch size
    return image

def predict_image(image_path):
    image = preprocess_image(image_path)
    prediction = model.predict(image)
    class_label = np.argmax(prediction)
    return class_label

# Define the class labels
class_labels = {0: 'Bacteria', 1: 'Fungi', 2: 'Nematodes', 3: 'Normal', 4: 'Virus'}

# Example usage
image_path = './Datasets/Virus/02e6c80d-c86f-44ca-9d4c-4b100f92a839___RS_GLSp 4631.JPG'
predicted_class = predict_image(image_path)

# Display the image and predicted class
image = Image.open(image_path)
plt.imshow(image)
plt.title(f"Predicted : {class_labels[predicted_class]}")
plt.axis('off')
plt.show()
```
